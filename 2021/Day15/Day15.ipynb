```julia
import Pkg; Pkg.add("DataStructures"); using DataStructures;
```

```julia
function init()
    return [[parse(Int64, d) for d in l] for l in readlines("input.txt")]
end;
```

```julia
struct PathNode
    coord::Tuple{Int64, Int64}
    prev::Union{PathNode, Nothing}
end;
```

```julia
@time begin

map = init();
start_coord = (1,1);
end_coord = (length(map), length(map[1]));

x_range = 1:end_coord[1];
y_range = 1:end_coord[2];
in_bounds(t::Tuple{Int64, Int64}) = (t[1] in x_range) && (t[2] in y_range);
deltas = [(-1,0),(1,0),(0,-1),(0,1)];
 
start = PathNode(start_coord, nothing);
priority_queue = PriorityQueue{PathNode, Int64}([]);
visited = Set{Tuple{Int64, Int64}}([]);

current = start;
current_dist = 0;
push!(visited, start_coord);
while current.coord != end_coord
    for delta in deltas
        next_coord = current.coord .+ delta;
        if in_bounds(next_coord) && !(next_coord in visited)
            enqueue!(priority_queue, PathNode(next_coord, current), current_dist + map[next_coord[1]][next_coord[2]])
            push!(visited, next_coord);
        end;
    end;
    current, current_dist = peek(priority_queue);
    dequeue!(priority_queue);
end;

println("min risk: ", current_dist);
end
```

```julia
@time begin

map = init();
map_x_dim = length(map);
map_y_dim = length(map[1]);

start_coord = (1,1);
end_coord = (map_x_dim * 5, map_y_dim * 5);
map_size = end_coord[1] * end_coord[2];

x_range = 1:end_coord[1];
y_range = 1:end_coord[2];
in_bounds(t::Tuple{Int64, Int64}) = (t[1] in x_range) && (t[2] in y_range);
deltas = [(-1,0),(1,0),(0,-1),(0,1)];

function get_risk(t::Tuple{Int64, Int64})
    x_wraps = div(t[1] - 1, map_x_dim);
    local_x = mod1(t[1], map_x_dim);
    y_wraps = div(t[2] - 1, map_y_dim);
    local_y = mod1(t[2], map_y_dim);
    return mod1(map[local_x][local_y] + x_wraps + y_wraps, 9);
end;
    
start = PathNode(start_coord, nothing);
priority_queue = PriorityQueue{PathNode, Int64}([]);
visited = Set{Tuple{Int64, Int64}}([]);

current = start;
current_dist = 0;
push!(visited, start_coord);
i = 0;
while current.coord != end_coord
    for delta in deltas
        next_coord = current.coord .+ delta;
        if in_bounds(next_coord) && !(next_coord in visited)
            enqueue!(priority_queue, PathNode(next_coord, current), current_dist + get_risk(next_coord))
            push!(visited, next_coord);
        end;
    end;
    current, current_dist = peek(priority_queue);
    dequeue!(priority_queue);
    i += 1;
    if i > 10000
        i = 0;
        println("visited: ",length(visited),"/",map_size);
    end;
end;
println("visited: ",length(visited),"/",map_size);
    
println("min risk: ", current_dist);
end
```
